```python
# prime number 
x = int(input('Enter number'))
i = 2
while i < x:
    if x%i==0:
      print(f'Not prime because it is divisible by {i}')
      break
    i += 1
else:
   print(f'{x} is prime Number')          
```

```python
# list 

def list_go(list_):
    list_.append(1) 

x = [1,2,3]
# here we have not stored changes but still list changed 
list_go(x)
print(x)
```

```python
# dict mutable 
def dict_go(dict_):
    # 1st way to insert in dict 
    dict_.update({"namaste":3})
    # 2nd way to insert in dict
    dict_['salam'] = 4 
    
go = {"hello":1,"hey":2}
# same like list mutable 
dict_go(go)
print(go)
```

```python
#----reverse dictonary------#
 
dictinory = {'hello': 1, 'hey': 2, 'namaste': 2, 'salam': 4}
# reverse replaced key<->value 
reverse = {v:k for k , v in dictinory.items()}

# -----reverse dictonary 2 way------#
second_way = [k for k , v in dictinory.items() if v==2 ][0]
print(second_way)
```

```python
# ----- append / extend ------

x = [1,2]
x.append([2,3]) # append instert as it is  
print(x) 
y = [2,3]
y.extend([3,4]) # extend instert one by one (deep down it follows append with loop for len of insert value )
print(y)
```

```python
# ---- int / str / float / tuples [IMMUTABLE] ----

def go(x):
    x += 4
    return x

x = 4 # can be str , float 
go(x) # we have to store function o/p inside other variable cuz int is immutable 
print(x) 

tuple_ = (1,2)
print(tuple_) # tuple doest not have any append function
```

```python
# ----how to preform insertion in tuple ----
tuple_go = (1,2,3)
# first trick -> change list into tuple then again into tuple 
tuple_go = list(tuple_go)
tuple_go.append(4)
tuple_go = tuple(tuple_go)
print(tuple_go)
# second way is concadination 
tuple_go = tuple_go + (5,)
print(tuple_go)
# for specific position
tuple_go = tuple_go[:1] + (6,) + tuple_go[1:]
print(tuple_go)
```

```python
# tuple loop 
for i in range(len(tuple_go)):
    if tuple_go[i] == 7:
        print('element found')
        break
else:    
    print('not found ')    
```
